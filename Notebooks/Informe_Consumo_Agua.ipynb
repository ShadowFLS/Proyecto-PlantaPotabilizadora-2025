# Análisis del Consumo de Agua Potable – Versión Completada

Este informe utiliza la versión final del dataset con valores faltantes completados mediante métodos estadísticos.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar dataset completo
df = pd.read_excel("Consumo_Agua_Enriquecido_COMPLETADO.xlsx")

# Asegurar tipos correctos
df["VARIACION MENSUAL (%)"] = pd.to_numeric(df["VARIACION MENSUAL (%)"], errors="coerce")
```

## Evolución del Consumo Total Mensual

```python
plt.figure(figsize=(10, 6))
plt.plot(df["MES"], df["TOTAL MENSUAL"], marker='o', linewidth=2)
plt.title("Evolución del Consumo Total Mensual de Agua")
plt.xlabel("Mes")
plt.ylabel("Litros consumidos")
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Consumo Promedio Anual por Barrio

```python
barrios = df.columns[1:14]  # Excluye MES, TOTAL, etc.
promedios = df[barrios].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
promedios.plot(kind='bar')
plt.title("Consumo Promedio Anual por Barrio (Litros)")
plt.ylabel("Litros promedio")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

## Variación Mensual Porcentual del Consumo

```python
plt.figure(figsize=(10, 5))
plt.plot(df["MES"], df["VARIACION MENSUAL (%)"], marker='s', color='orange')
plt.title("Variación Mensual Porcentual del Consumo")
plt.xlabel("Mes")
plt.ylabel("Variación (%)")
plt.axhline(0, color='gray', linestyle='--')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
